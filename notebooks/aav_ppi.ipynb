In conclusion, we **should not** use calculations based on centre of mass. This is expected anyway since we are not interested in the exact masses of the beads to begin with in MARTINI force field. This is because MARTINI force field ascribes 72 amu for all beads, except the ring beads which have 45 amu, for the sake of computational efficiency.
