## Orientational analysis for AAV2 capsid in the membrane environment

```python
import numpy as np
import MDAnalysis as mda
import aav_aavr_interaction
```

```python
'''
Orientation Analysis of AAV2 capsid will focus on the vector whose base is at the BB bead of resid 18580 and ends at BB bead of resid 13000.
We will use dot product to calculate how much this vector rotates over time.

This could also be used to calculate how much the diameter of the capsid changes over time.
'''

u = mda.Universe('/Users/pp420/mres_aav_aavr_project/aav2_aavr_scaled_3/dir_v2r1/v2r1_nosolv.gro', 
                 '/Users/pp420/mres_aav_aavr_project/aav2_aavr_scaled_3/dir_v2r1/v2r1_nosolv.xtc')

data = aav_aavr_interaction.assemble_timeseries_array(u=u,selection="(name BB and resid 13000) or (name BB and resid 18580)")

vectors = data[:,1,:3] - data[:,0,:3]
ref = vectors[0]
cosines = np.zeros(len(vectors))
norms = np.zeros(len(vectors))
for i in range(len(vectors)):
    c = np.dot(vectors[i],ref)/(np.linalg.norm(vectors[i])*np.linalg.norm(ref))
    cosines[i] += c
    norms[i] += np.linalg.norm(vectors[i])

np.save("orientation/v2r1_cos.npy", cosines)
np.save("orientation/v2r1_norms.npy", norms)

u = mda.Universe('/Users/pp420/mres_aav_aavr_project/aav2_aavr_scaled_3/dir_v2r2/v2r2_nosolv.gro', 
                 '/Users/pp420/mres_aav_aavr_project/aav2_aavr_scaled_3/dir_v2r2/v2r2_nosolv.xtc')

data = aav_aavr_interaction.assemble_timeseries_array(u=u,selection="(name BB and resid 13000) or (name BB and resid 18580)")

vectors = data[:,1,:3] - data[:,0,:3]
ref = vectors[0]
cosines = np.zeros(len(vectors))
norms = np.zeros(len(vectors))
for i in range(len(vectors)):
    c = np.dot(vectors[i],ref)/(np.linalg.norm(vectors[i])*np.linalg.norm(ref))
    cosines[i] += c
    norms[i] += np.linalg.norm(vectors[i])

np.save("orientation/v2r2_cos.npy", cosines)
np.save("orientation/v2r2_norms.npy", norms)

u = mda.Universe('/Users/pp420/mres_aav_aavr_project/aav2_aavr_scaled_3/dir_v2r3/v2r3_nosolv.gro', 
                 '/Users/pp420/mres_aav_aavr_project/aav2_aavr_scaled_3/dir_v2r3/v2r3_nosolv.xtc')

data = aav_aavr_interaction.assemble_timeseries_array(u=u,selection="(name BB and resid 13000) or (name BB and resid 18580)")

vectors = data[:,1,:3] - data[:,0,:3]
ref = vectors[0]
cosines = np.zeros(len(vectors))
norms = np.zeros(len(vectors))
for i in range(len(vectors)):
    c = np.dot(vectors[i],ref)/(np.linalg.norm(vectors[i])*np.linalg.norm(ref))
    cosines[i] += c
    norms[i] += np.linalg.norm(vectors[i])

np.save("orientation/v2r3_cos.npy", cosines)
np.save("orientation/v2r3_norms.npy", norms)
```
